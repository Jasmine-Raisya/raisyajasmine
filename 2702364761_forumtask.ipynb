```python
lloyd = {
    "name": "Lloyd",
    "homework": [90.0, 97.0, 75.0, 92.0],
    "quizzes": [88.0, 40.0, 94.0],
    "tests": [75.0, 90.0]
}
alice = {
    "name": "Alice",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0]
}
tyler = {
    "name": "Tyler",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0]
}

students = [lloyd,alice,tyler]

# Define the get_average function to return the average
def get_average(student):
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])
    # Return the average as a result of this function
    return 0.1 * homework + 0.3 * quizzes + 0.6 * tests

def average(numbers):
    tot = sum(numbers)
    avg = float(tot)
    return avg / len(numbers)

def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Calculate the average and print the letter grade
print(get_letter_grade(get_average(lloyd)))

def get_class_average(students):
    result = []

    for students in students:
        student_average = get_average(students)
        result.append(student_average)
    return average(result)

print(get_class_average(students))
print(get_letter_grade(get_class_average(students)))
   
```

```python
week_days = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
sales = []

for day in week_days:
    while True:
        sales_ = int(input(f"Enter the sales for {day}:"))
        if sales_ >=0:
            sales.append(sales_)
            break
        else:
            print(f"This input is invalid.")

sum = sum(sales)
minim = min(sales)
maxim = max(sales)

print(f"The total sales is: ${sum:,.2f}")
print(f"The minimum sale amount was: ${minim:,.2f}")
print(f"The maximum sale amount was: ${maxim:,.2f}")
```
